```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
def drow_aplt(data ,ii,name):
    ns = []
    for i in range(min(data.index),max(data.index),6):
        sum = 0
        for j in range(6):
            if i + j == len(data.index):
                break
            sum += data['n'][j+i]
        ns.append(sum)
    beta = max(ns)/max(max(data['var_distance']),max(data['distance']))
    ns = np.array(ns)/beta
    if name == 'Unknown':
        plt.subplot(5,4,ii)
        plt.title(name)
        plt.bar(1993,ns, color='blue',alpha = 0.2, width=5)
        plt.plot(data['year'].values,data['distance'].values,c='r')
        plt.plot(data['year'].values,data['var_distance'].values,c = 'black')
        return
    plt.subplot(5,4,ii)
    plt.title(name)
    plt.bar(np.linspace(data['year'][0]+3, data['year'][max(data.index)]-3,len(ns)),ns, color='blue',alpha = 0.2, width=5)
    plt.plot(data['year'].values,data['distance'].values,c='r')
    plt.plot(data['year'].values,data['var_distance'].values,c = 'black')
    # plt.show()
    # plt.savefig('%s.jpg'%(name.replace('/','_')),dpi=200)
    # plt.clf()
```

```python
plt.figure(figsize=(16,12))
plt.subplots_adjust(wspace =0.2, hspace =0.4)
sheets = pd.read_csv('../art_genre.tsv',sep='\t')['genre'].unique()
for i in  range(len(sheets)):
    data = pd.read_excel('music_var_year_genre_pca3.xlsx',sheet_name=sheets[i].replace('/','_'))
    drow_aplt(data,i+1,sheets[i])
# plt.legend()
plt.savefig('feature_time.jpg',dpi=200)
```
