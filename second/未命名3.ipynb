```python
import pandas as pd
```

```python
data = pd.read_csv('data/full_music_data.csv')
data = data.sort_values('year')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,12))
plt.subplots_adjust(wspace = 0.4, hspace = 0.4)
# fig, ax = plt.subplots()
columns = data.columns[2:-2]
for i in range(len(columns)):
    x = data[columns[i]]
    n = len(x)
    plt.subplot(4,4,i+1)
    plt.title(columns[i])
    if len(x.unique()) <= 20:
        plt.hist(x.values)
    else:
        plt.scatter(range(n),x.values,s=0.02,alpha=0.8)
plt.savefig('stastic.jpg')
```
