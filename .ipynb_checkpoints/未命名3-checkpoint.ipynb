```python
import pandas as pd
```

```python
data = pd.read_csv('data/full_music_data.csv')
data
```

```python
data.columns[2:-3]
```

```python
len(data.columns)
```

```python
import matplotlib.pyplot as plt
plt.figure((12,8))
columns = data.columns[2:-3]
for i in range(len(columns)):
    x = data[columns[i]]
    n = len(x)
    plt.subplot(4,4,i+1)
    if len(x.unique()) <= 20:
        plt.hist(x.values)
    else:
        plt.scatter(range(n),x.values)
plt.show()
```
