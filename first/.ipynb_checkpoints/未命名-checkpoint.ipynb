```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 读入数据
data = pd.read_csv('../data/influence_data.csv')
data
```

```python
# 生成邻接矩阵
n = max(max(data['inf_id'].values)+1, max(data['flw_id'].values)+1)
A = np.zeros((n, n))
for i in range(len(data)):
    A[data['flw_id'][i], data['inf_id'][i]] = 1
A
```

## 聚合系数

```python
# 计算聚合系数 ,存入clst数组
clst = []
for i in range(len(A)):
    k = sum(A[i, :]) + sum(A[:, i])
    print(k)
    e = 0
    neighbor = []
    for j in range(len(A)):
        if A[i, j] == 1 or A[j, i] == 1:
            neighbor.append(j)
    # set用来选出不重复的相邻节点
    neighbor = set(neighbor)
    for j in neighbor:
        for l in neighbor:
            e += A[j, l]+A[l, j]
    if k != 0 and k != 1:
        clst.append(2/k/(k-1)*e)
    else:
        clst.append(0)
```

```python
sorted(clst, reverse=True)
```

```python
# 存储每一节点追随的人
mmp = {}
for i in range(len(A)):
    mmp[i] = np.where(A[:, i])
mmp
```

```python
# 带有深度限制的dfs, 寻找叶子节点(入度为0)
# fuck数组避免成环
def dfs(node, ttl):
    global fuck
    fuck.append(node)
    if ttl == 0:
        return 0
    sum = 0
    if len(mmp[node][0]) == 0:
        return 0
    for i in mmp[node][0]:
        if i in fuck:
            continue
        sum += dfs(i, ttl-1) + 1
    return sum
```

```python
ccc = {}
for i in range(len(A)):
    fuck = []
    ccc[i] = dfs(i, 100)
    print(i)
```

```python
ccc
```

```python
# 计算特征路径长度
shit = {}
for i in ccc.keys():
    k = sum(A[:, i])
    if k == 0:
        shit[i] = 0
    else:
        shit[i] = ccc[i]/k
print(shit)
```

## katz 中心性

```python
# 求特征值
lmds, _ = np.linalg.eig(A)
lmd = np.max(lmds)
lmd
```

```python
# 根据特征值估算, a取1/4
a = 1/4
Ik = np.zeros(A.shape)
pre_A = np.identity(len(A))
for i in range(100):
    # 迭代, 考虑时间复杂度, 只设置100次
    pre_A = a * pre_A.dot(A)
    Ik += pre_A
Ik
```

```python
# 存储影响别人的人和被影响的人的katz中心值
f1 = open('influcing_katz1.csv', 'a+')
f2 = open('influced_katz1.csv', 'a+')
for i in range(len(A)):
    f1.write("%d, %f\n" % (i, sum(Ik[:, i])))
    f2.write("%d, %f\n" % (i, sum(Ik[i, :])))
f1.close()
f2.close()
```
